```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("Life expectancy.csv")
```

```python
fig, axes = plt.subplots(nrows=5, ncols=3, sharex=True, sharey=True, figsize=(15,15))
axes_list = [item for sublist in axes for item in sublist] 

for countryname, selection in df.groupby("Entity"):
    ax = axes_list.pop(0)
    selection.plot(x='Year', y='Life expectancy', label=countryname, ax=ax, legend=False, clip_on=False)
    ax.set_title(countryname)
    ax.tick_params(
        which='both',
        bottom='off',
        left='off',
        right='off',
        top='off'
    )
    ax.grid(linewidth=0.25)
    ax.set_xlim((1800, 2016))
    ax.set_xlabel("")
    ax.set_xticks(range(1800, 2016, 50))
    ax.spines['left'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

plt.tight_layout()
fig.savefig("life_expectancy.jpg")
```
